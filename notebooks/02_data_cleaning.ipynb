```python
import pandas as pd
```

```python
df = pd.read_csv("../data/insurance_data.csv")
df.head()
```

### CLEAN the data

```python
clean_df = df.copy()
```

```python
#Remove duplicates
clean_df = clean_df.drop_duplicates()
```

```python
#Handle missing values
clean_df = clean_df.dropna()
```

```python
#Fix column names
clean_df.columns = clean_df.columns.str.lower().str.strip()
```

```python
#Convert dates
clean_df["transactiondate"] = pd.to_datetime(clean_df["transactiondate"])
```

```python
#SAVE CLEAN DATASET
clean_df.to_csv("../data/insurance_data_cleaned.csv", index=False)
```
